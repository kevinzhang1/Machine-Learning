```python
#Import the dataset
import io
from google.colab import files
uploaded = files.upload()
dataset=pd.read_csv(io.BytesIO(uploaded['Mall_Customers.csv']))
#dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

```python
# P5. 1 Use the elbow method to find the optimal number of clusters through a loop (1 to 10 clusters) and visualize the result on a 2D plot
# Set init = 'k-means++'
a=[]
for i in range(1, 10):
    k=KMeans(n_clusters=i, init='k-means++')
    k.fit(X)
    a.append(k.inertia_)
import seaborn as sns
sns.lineplot(range(1, 10), a)
plt.show()
```

```python
# P5. 2 Fit K-Means to the dataset and use the optimal number of clusters that you found in the previous part and show the result
k.fit(X)
print(k)
```

```python
# P5. 3 Visuale the clusters (# xlabel('Annual Income (k$)' and ylabel('Spending Score (1-100)'))
y_kmeans = k.fit_predict(X)
# Use different colors for clusters and label them
plt.scatter(X[y_kmeans==0, 0], X[y_kmeans==0, 1], c='brown')
plt.scatter(X[y_kmeans==1, 0], X[y_kmeans==1, 1], c='black')
plt.scatter(X[y_kmeans==2, 0], X[y_kmeans==2, 1], c='orange')
plt.scatter(X[y_kmeans==3, 0], X[y_kmeans==3, 1], c='blue')
plt.scatter(X[y_kmeans==4, 0], X[y_kmeans==4, 1], c='green')
plt.xlabel('Annual Income(k$)')
plt.ylabel('Spending Score(1-100')
plt.show()
```
